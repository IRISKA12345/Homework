```python
def bin_search(a, x):
    left = 0
    right = len(a) - 1
    while left <= right:
        mid = (left + right) // 2

        if a[mid] == x:
            return mid
        elif a[mid] < x:
            left = mid + 1
        else:
            right = mid - 1

    return -1

a = [2, 3, 5, 1, 9]
print(bin_search(a, 3))
```

```python
 def gcd(a, b):
  if a == 0:
    return b
  if b== 0:
    return a
  else:
    return gcd(b, a % b)

print(gcd(3, 5))
```

```python
n = int(input())
def hanoi_tower( a, b, c, n):
  if n == 1:
    print(a, "->", b)
  else:
    hanoi_tower(a, c, b, n-1)
    print(a, "->", b)
    hanoi_tower(c, b, a, n-1)
```

```python
def print_triangle(size, symbol):
  if size % 2 == 0:
    peak = size // 2
  else:
    peak = size // 2 + 1
  for i in range(1, peak + 1):
    print(symbol * i)
  for i in range(peak - 1, 0, -1):
    print(symbol * i)
size = int(input())
symbol = input()
print({size})
print_triangle(size, symbol)
```

```python
import string

def count_words(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        text = file.read()

    translator = str.maketrans(string.punctuation, ' ' * len(string.punctuation))
    cleaned_text = text.translate(translator)
    cleaned_text = cleaned_text.lower()

    words = cleaned_text.split()

    word_count = {}

    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    return word_count

def find_top_10(word_count):
    top_words = []

    temp_dict = word_count.copy()

    for i in range(10):
        max_word = None
        max_count = 0

        for word, count in temp_dict.items():
            if count > max_count:
                max_count = count
                max_word = word

        if max_word:
            top_words.append((max_word, max_count))
            del temp_dict[max_word]

    return top_words

filename = "инфа.txt"

word_count = count_words(filename)

top_10 = find_top_10(word_count)

for i, (word, count) in enumerate(top_10, 1):
    print(f"{i:2}. {word:15} - {count:3} раз(a)")
```
